```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
def display_header_tags(url):
    
    page = requests.get(url)
    print(page,"\n")
    soup = BeautifulSoup(page.content,'lxml')
    
    print("Header tags from the website are as follows:","\n")
    for i in soup.find_all(['h1','h2','h3','h4','h5','h6']):
        print(i,"\n")
```

Q2) 
Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame

```python
def top_movie_df(url):
    url = "https://www.imdb.com/chart/top/"
    page = requests.get(url)
    
    soup = BeautifulSoup(page.content, "lxml")
    
    parent = soup.find_all('td', {'class' : 'titleColumn'})[0:100]
    #HERE we create a list of MOVIE NAMES from MOST POPULAR MOVIES
    
    title = []
    for item in parent:
        title.append(item.find('a').text)
    #HERE we create a list of MOVIE release years   
    
    year=[]
    for i in parent:
        year.append(i.find('span').text[1:5])
    #HERE we create a list of MOVIE ratings   
    
    parent2 = soup.find_all("td",class_="ratingColumn imdbRating")[0:100]
    rating=[]
    for i in parent2:
        rating.append(i.find("strong").text)
        
    #create a data frame
    df=pd.DataFrame({"Movie name": title, "Year of release": year, "Imdb Rating": rating})
    return df
```

```python
top_movie_df("https://www.imdb.com/chart/top/")
```

```python
#url in all the parts of question 4 and 5 is different
def top_odi_teams(url):
    url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "lxml")
    team_name=[]
    parent = soup.find_all("span",class_= "u-hide-phablet")[0:10]
    for item in parent:
        team_name.append(item.text)
    team1_matches = soup.find("td",class_="rankings-block__banner--matches").text
    team1_points =  soup.find("td",class_="rankings-block__banner--points").text
    team1_ratings = soup.find("td",class_="rankings-block__banner--rating u-text-right").text.strip()
    team_matches_left = []
    remaining_team_matches=soup.find_all("td",class_="table-body__cell u-center-text")[0:18:2]
    for i in remaining_team_matches:
        team_matches_left.append(i.text)
    team_matches_left.insert(0,team1_matches)
    team_points = []
    x=soup.find_all("td",class_="table-body__cell u-center-text")[1:18:2]
    for i in x:
        team_points.append(i.text)
    team_points.insert(0,team1_points)
    team_ratings = []
    x=soup.find_all("td",class_="table-body__cell u-text-right rating")[0:9]
    for i in x:
        team_ratings.append(i.text)
    team_ratings.insert(0,team1_ratings)
    df = pd.DataFrame({"Team name": team_name,"Matches":team_matches_left,"Points":team_points,"Ratings": team_ratings})
    return df
```

Q6) Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def phones_under20k(url):
    url = "https://www.amazon.in/best-mobile-under-20000/s?k=best+mobile+under+20000"
    page = requests.get(url)
   
    soup = BeautifulSoup(page.content, "lxml")
    
    phone1 = soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")#
    phone_name = []
    for i in phone1:
        phone_name.append(i.text.partition(" (")[0])
    
    rating1 = soup("div", class_="a-section a-spacing-none a-spacing-top-micro")[0:32:2]
    ratings=[]
    for i in rating1:
        ratings.append(i.find("span").get('aria-label').replace("stars",""))
   
    img1 = soup("div",class_="a-section aok-relative s-image-fixed-height")
    link=[]
    for i in img1:
        link.append(i.find("img").get('src'))

    price1 = soup.find_all("span", class_="a-price-whole")
    prices=[]
    for i in price1:
        prices.append(i.text)
    
    df = pd.DataFrame({"Phone Name": phone_name,"Price":prices,"Rating": ratings,"Image link": link})
    return df
    
```

```python
phones_under20k("https://www.amazon.in/best-mobile-under-20000/s?k=best+mobile+under+20000")
```

```python
def properties(url):
    url = "https://www.nobroker.in/property/sale/bangalore/Electronic%20City?type=BHK4&searchParam=W3sibGF0IjoxMi44NDUyMTQ1LCJsb24iOjc3LjY2MDE2OTUsInBsYWNlSWQiOiJDaElKdy1GUWQ0cHNyanNSSGZkYXpnXzhYRW8iLCJwbGFjZU5hbWUiOiJFbGVjdHJvbmljIENpdHkifV0=&propertyAge=0&radius=2.0"
    page = requests.get(url)

    soup = BeautifulSoup(page.content, "lxml")
    location = soup.find_all('h2',class_="heading-6 font-semi-bold nb__1AShY")
    title = []
    addr= []

    for i in location:
        title.append(i.find("span").text.partition(" For Sale  In ")[0].replace(" ",""))
        addr.append(i.find("span").text.partition(" For Sale  In ")[2].replace(", ",""))
    all_info = []
    emi=[]
    area=[]
    for i in soup.find_all("div",class_="nb__2NPHR"):
        all_info.append(i.text)
    del all_info[0::4]
    area = all_info[0::3]
    emi = all_info[1::3]
    correct_emi=[]
    for i in emi:
        correct_emi.append(i.replace("Estimated EMI","").replace("₹",""))
    price = all_info[2::3]
    final_price=[]
    for i in price:
        final_price.append(i.replace("₹","").partition("s")[0])
    df = pd.DataFrame({"Prop title": title,"Location": addr,"Area":area, "Price":final_price,"EMI": correct_emi})
    aligned_df = df.style.set_properties(**{'text-align': 'left'})
    return aligned_df
```

```python
def dineout(url):
    url = "https://www.dineout.co.in/delhi-restaurants/buffet-special"
    page = requests.get(url)

    soup = BeautifulSoup(page.content, "lxml")

    name = soup.find_all('div',class_="restnt-info cursor")
    restro_name = []
    for i in name:
        restro_name.append(i.find("a").text)
    
    loc = soup.find_all("div",class_="restnt-loc ellipsis")
    location = []
    for i in loc:
        location.append(i.text)
    
    price = []
    for i in soup.find_all("div",class_="detail-info"):
        price.append(i.text.partition(" |")[0].replace("₹",''))

    rating = []
    for i in soup.find_all("div",class_="img-wrap"):
        rating.append(i.text)
    rating[2] = rating[2].replace("In high demand",'')
    
    cuisine=[]
    for i in soup.find_all("div",class_="detail-info"):
        cuisine.append(i.text.partition(" |")[2].replace("1 offer available",""))
    url = []
    for i in soup.find_all("div",class_="img-wrap"):
        url.append(i.find("img",class_="no-img").get("data-src"))
    
    df = pd.DataFrame({"Restaurant Name": restro_name,"Location": loc, "Price":price,"Rating": rating,"Cuisine":cuisine ,"Image link": url})
    aligned_df = df.style.set_properties(**{'text-align': 'left'})
    return aligned_df
    
```

```python
def weather_24_last(url):
    page = requests.get("https://en.tutiempo.net/delhi.html?data=last-24-hours")
    soup = BeautifulSoup(page.content, "lxml")
    Hour = []

    for i in soup.find_all('td',class_='t Temp')[0:48]: 
        if i.previous_sibling is not None:
            Hour.append(i.previous_sibling.previous_sibling.text)
        else:
            Hour.append('')
    temp=[]
    for i in soup.find_all ('td',class_='t Temp')[0:48]: 
        temp.append(i.text)
    Weather_condition=[]
    for i in soup.find_all ('td',class_='t Temp')[0:48]: 
        if i.previous_sibling is not None:
            Weather_condition.append(i.previous_sibling.text)
        else:
            Weather_condition.append(' ')
    wind=[]
    for i in soup.find_all ('td',class_='wind')[0:48]: 
        wind.append(i.text)
    hum=[]
    for i in soup.find_all ('td',class_='hr')[0:48]: 
        hum.append(i.text)
    pres=[]
    for i in soup.find_all ('td',class_='prob')[0:48]: 
        pres.append(i.text)
    df = pd.DataFrame({"Hour":Hour,"Temp":temp,"Wind": wind,"WC": Weather_condition, "humidity": hum, "Pressure":pres})
    aligned_df = df.style.set_properties(**{'text-align': 'left'})
    return aligned_df
weather_24_last("https://en.tutiempo.net/delhi.html?data=last-24-hours")
```

Q10) Write a python program to scrape monument name, monument description, image url about top 10 monuments from 'https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/'

```python
def top_10_monu(url):
    page = requests.get("https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/")
    soup = BeautifulSoup(page.content, "lxml")
    data = soup.find_all('div', class_="blog--single__content column--3-4 u-spacing-third")
    temp=[]
    for i in data:
        for j in i.find_all('p'):
            t = j.text.replace('\n', '') # only extracting data
            temp.append(t)
    del temp[-3:]
    temp.pop(0)
    del temp[2::3]
    name = temp[0::2]
    descr = temp[1::2]
    url=[]

    for d in data:
        for s in d.find_all('p'):   #img tag is inside a p tag, so well have to iterate over all of them
            for m in s.find_all("img"):  # to extract img data
                l = m.get("src") 
                url.append(l)
    urls = url[1::2]
    df = pd.DataFrame({"Name":name,"Description":descr,"Image URL": urls})
    aligned_df = df.style.set_properties(**{'text-align': 'left'})
    return aligned_df
top_10_monu("https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/")
```
