2. Scrape the details team India’s international fixtures from bcci.tv.

You need to find following details:
- Match title (I.e. 1st ODI)
- Series
- Place
- Date
- Time

Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
url = "https://www.bcci.tv/."
driver.get(url)
wait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, "//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover'][1]")))
hover1 = driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover'][1]")
hover1.click()
sleep(0.3)
hover2=driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down'][1]")
hover2.click()
sleep(0.5)
wait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, "//span[@class='button__text']")))
btn_click= driver.find_element(By.XPATH, "//span[@class='button__text']").click()
```

```python
title=[]
series=[]
place=[]
date_time=[]

datetags=driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']")
for tag in datetags:
    try:
        date_time.append(tag.text. replace("\n"," "))
        
    except NoSuchElementException:
        date_time.append("-")
    

seriestags=driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
for tag in seriestags:   
    try:
        series.append(tag.text)
    except NoSuchElementException:
        series.append("-")

titletags=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']//strong")
for tag in titletags:
    try:
        title.append(tag.text)
    except NoSuchElementException:
        title.append("-")

placetags=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']//span[1]")
for tag in placetags:
    try:
        place.append(tag.text)
    except NoSuchElementException:
        place.append("-")


pd.DataFrame({"Date and Time":date_time,"Title":title,"Series":series,"Place":place})
```

```python
title=[]
descr=[]
lang=[]
contr=[]

#title 
try:
    titletags=driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/h1/a")
    for tag in titletags:
          title.append(tag.text)
except NoSuchElementException:
          title.append("-")

#language
for i in range(1,len(title)+1):
    try:
        lang_name=driver.find_element_by_xpath(f"//*[@id='js-pjax-container']/div[3]/div/div[2]/article[{i}]/div[2]/span[1]/span[2]")
        lang.append(lang_name.text)
    except:
        lang.append("-")


#description
try:
    desctags=driver.find_elements_by_xpath("//article[@class='Box-row']//p")
    for i in desctags:
          descr.append(i.text)
except NoSuchElementException:
          descr.append("-")



#contributors
for i in range(1,len(title)+1):
    try:
        contr_name=driver.find_element_by_xpath(f"//*[@id='js-pjax-container']/div[3]/div/div[2]/article[{i}]/div[2]/a[2]")
        contr.append(contr_name.text)
    except:
        contr.append("-")

        
```

```python
print("titles length",len(title))
print("Description length",len(descr))
print("Language",len(lang))
print("Number of Contributors",len(contr))
```

```python
name=[]
year=[]
runt=[]
stars=[]
genre=[]
votes=[]

nametags=driver.find_elements_by_xpath("//h3[@class='lister-item-header']//a")
for tag in nametags:
    try:
        name.append(tag.text)
    except NoSuchElementException:
        name.append('-')  
        

#range
yeartags=driver.find_elements_by_xpath("//*[@id='main']/div/div[3]/div[3]/div/div[2]/h3/span[2]")
for tag in yeartags:
    try:
        year.append(tag.text)
    except NoSuchElementException:
        year.append('-')
        
#genre
gtags=driver.find_elements_by_xpath("//span[@class='genre']")
for tag in gtags:
    try:
        genre.append(tag.text)
    except NoSuchElementException:
        genre.append('-') 

#RunTime
runtags=driver.find_elements_by_xpath("//span[@class='runtime']")
for tag in runtags:
    try:
        runt.append(tag.text)
    except NoSuchElementException:
        runt.append('-') 
        
#ratings
startags=driver.find_elements_by_xpath("//span[@class='ipl-rating-star__star']//span")
for tag in startags:
    try:
        stars.append(tag.text)
    except NoSuchElementException:
        stars.append('-') 
#votes
votetags=driver.find_elements_by_xpath("//div[@class='lister-item mode-detail']/div[2]/p[4]/span[2]")
for tag in votetags:
    try:
        votes.append(tag.text)
    except NoSuchElementException:
        votes.append('-') 
```
