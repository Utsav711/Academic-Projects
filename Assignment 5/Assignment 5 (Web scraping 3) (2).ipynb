1. Write a python program which searches all the product under a particular product vertical
from www.amazon.in. The product verticals to be searched will be taken as input from user.
For e.g. If user input is ‘guitar’. Then search for guitars.

2. In the above question, now scrape the following details of each product listed in first 3 pages
of your search results and save it in a data frame and csv. In case if any product vertical has
less than 3 pages in search results then scrape all the products available under that product
vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details"
and “Product URL”. In case, if any of the details are missing for any of the product then
replace it by “-“.

```python
url = "https://www.amazon.in/"
driver.get(url)
sleep(4)

item = input("Enter item you want to search: ")
search_prod = driver.find_element_by_xpath("//div[@class='nav-search-field ']/input[1]")
search_prod.send_keys(item)
search_btn = driver.find_element_by_xpath("//span[@id='nav-search-submit-text']/input[1]")
search_btn.click()
sleep(3)
```

```python
product_headers=[]
stars=[]
No_ratings=[]
price_list=[]
Expect=[]
available=[]
producturl=[]
exchange=[]
```

```python
driver.switch_to.window(driver.window_handles[0]) #switch back to original tab
products = driver.find_elements(By.XPATH, "//a[@class='a-link-normal a-text-normal']")
print(len(products))
for product in products[0:5]:
    
#     driver.get(product)
    sleep(1)
#     driver.execute_script("arguments[0].scrollIntoView();", product)
    driver.execute_script("window.scrollBy(0,150)")
    product.click()
    driver.switch_to.window(driver.window_handles[-1]) #switch to new tab
    sleep(2)
#     wait(driver, 5).until(EC.element_to_be_clickable((By.ID, 'titleSection')))
    try:
        header_details = driver.find_element(By.XPATH, '''//*[@id="titleSection"]''') #get product header text
        product_headers.append(header_details.text)
    except:
         product_headers.append("-")
      
    sleep(3)
#     wait(driver, 5).until(EC.element_to_be_clickable((By.XPATH,'''//td[@class='a-size-base']''')))
    
    try:        
        star = driver.find_element(By.XPATH,"//span[@class='a-size-medium a-color-base']")   #get product stars text
        stars.append(star.text)
    except:
        stars.append('-')
        
    try:
        Rating = driver.find_element(By.ID,'acrCustomerReviewText')
        No_ratings.append(Rating.text)
    except:
        No_ratings.append('-')

    try:
        delivery_date = driver.find_element(By.CSS_SELECTOR,'#ddmDeliveryMessage > b:nth-child(2)')   #get delivery date text
        Expect.append(delivery_date.text.split('\n')[0])
    except:
        Expect.append('-')
    
    try: 
        ex=driver.find_element_by_xpath("//div[@class='a-column a-span3 a-text-center icon-container icon-grid-width celwidget']").text
        exchange.append(ex)
    except:
        exchange.append("-")
        
    try:    
        price = driver.find_element(By.ID,"price_inside_buybox").text   #get product price text
        price_list.append(price)
    except:
        price_list.append('-')

    try:
        in_stock = driver.find_element(By.XPATH,"//span[@class='a-size-medium a-color-success']")   #get IN Stock text
        available.append(in_stock.text)
    except:
        available.append('-')

    url = driver.current_url #url of the webpage
    producturl.append(url)

    driver.close()                                                                  
    driver.switch_to.window(driver.window_handles[0]) #switch back to original tab
```

4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on
www.flipkart.com and scrape following details for all the search results displayed on 1st page.
Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”,
“Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display
Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”.
Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe
and CSV.


```python
def phone_specs(name):

    url = "https://www.flipkart.com/"
    driver.get(url)
    sleep(2)

    try:
        click_oncross = driver.find_element_by_xpath("//div[@class='_2QfC02']/button[1]")   
        click_oncross.click()                                                               
    except:
        pass

    search_prod = driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input[1]")
    search_prod.send_keys(name)
    search_btn = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
    search_btn.click()
    wait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, "//div[@class='_4rR01T']")))

    phone_tags= driver.find_elements(By.XPATH, "//div[@class='_4rR01T']")

    brand_name=[]
    phone_name=[]
    Color=[]
    ram=[]
    storage=[]
    primarycam=[]
    secondarycam=[]
    displaysize=[]
    displayres=[]
    processor=[]
    processor_cores=[]
    battery=[]
    price=[]
    product_url=[]

    for phone in phone_tags[0:3]:

        phone.click()
        driver.switch_to.window(driver.window_handles[-1]) #switch to new tab
        sleep(2)
        driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()   

        headers = driver.find_elements(By.XPATH, "//tr[@class='_1s_Smc row']")  #get all headers but dont search them

        header_data_dict = {} 
        for header in headers:  
            header_data_dict[header.text.split("\n")[0]]= header.text.split("\n")[1]

        try: 
            brand_name.append( driver.find_element(By.XPATH, "//span[@class='B_NuCI']").text.split()[0])
        except:
            brand_name.append("-")
        try: 
            phone_name.append(header_data_dict["Model Name"])
        except:
            phone_name.append("-")
        try: 
            Color.append(header_data_dict["Color"])
        except:
            Color.append("-")
        try: 
            displaysize.append(header_data_dict["Display Size"])
        except:
            displaysize.append("-")
        try: 
            displayres.append(header_data_dict["Resolution"])
        except:
            displayres.append("-")
        try: 
            processor.append(header_data_dict["Processor Type"])
        except:
            phone_name.append("-")
        try: 
            processor_cores.append(header_data_dict["Processor Core"])
        except:
            processor_cores.append("-")
        try: 
            storage.append(header_data_dict["Internal Storage"])
        except:
            storage.append("-")
        try: 
            ram.append(header_data_dict["RAM"])
        except:
            ram.append("-")
        try: 
            primarycam.append(header_data_dict["Primary Camera"])
        except:
            primarycam.append("-")
        try: 
            secondarycam.append(header_data_dict["Secondary Camera"])
        except:
            secondarycam.append("-")
        try: 
            battery.append(header_data_dict["Battery Capacity"])
        except:
            battery.append("-")
        try: 
            price.append( driver.find_element(By.XPATH, "//div[@class='_30jeq3 _16Jk6d']").text)
        except:
            price.append("-")
        try: 
            product_url.append( driver.current_url)
        except:
            product_url.append("-")

        driver.close()
        driver.switch_to.window(driver.window_handles[0]) #switch to new 

    df=pd.DataFrame({"Company name" : brand_name,
    "Model Name":phone_name,
    "Color":Color,
    "RAM":ram,
    "Internal Storage":storage,
    "Primary Camera":primarycam,
    "Secondary Cam":secondarycam,
    "Display Size":displaysize,
    "Display Resolution":displayres,
    "Processor":processor,
    "Processor cores":processor_cores,
    "Battery":battery,
    "Price":price,
    "Product URL":product_url})
    
    os.chdir(rf"C:\Users\utsav\Desktop\New folder")
    df.to_csv("phonespecs.csv")
```

Q6 :Write a program to scrap details of all the funding deals for second quarter (i.e. July 20 – September 20) from trak.in.

```python
url="https://trak.in/"

driver.get(url)

sleep(1)
driver .find_element_by_xpath("//a[@title='http://trak.in/india-startup-funding-investment-2015/']").click()
driver.execute_script("window.scrollTo(0, 6800)") 

date = []
start_name= []
industry= []
sub_vert= []
loc= []
invname= []
invtype = []
amt= []


for number in range(48,51):

    sub_tags = driver.find_elements_by_xpath(f'//table[@id="tablepress-{number}"]/tbody/tr/td[5]')
    for tag in sub_tags:
        sub_vert.append(tag.text)
        
    sname_tags= driver.find_elements_by_xpath(f'//table[@id="tablepress-{number}"]/tbody/tr/td[3]')
    for tag in sname_tags:
        start_name.append(tag.text)
    
    date_tags = driver.find_elements_by_xpath(f'//table[@id="tablepress-{number}"]/tbody/tr/td[2]')
    for tag in date_tags:
        date.append(tag.text)
        
    ind_tags = driver.find_elements_by_xpath(f'//table[@id="tablepress-{number}"]/tbody/tr/td[4]')
    for tag in ind_tags:
        industry.append(tag.text)
    
    amount_tags = driver.find_elements_by_xpath(f'//table[@id="tablepress-{number}"]/tbody/tr/td[9]/strong')
    for tag in amount_tags:
        amt.append(tag.text)
        
    inv_tags = driver.find_elements_by_xpath(f'//table[@id="tablepress-{number}"]/tbody/tr/td[7]')
    for tag in inv_tags:
        invname.append(tag.text)
    
    type_tags = driver.find_elements_by_xpath(f'//table[@id="tablepress-{number}"]/tbody/tr/td[8]')
    for tag in type_tags:
        invtype.append(tag.text)
    
    loc_tags = driver.find_elements_by_xpath(f'//table[@id="tablepress-{number}"]/tbody/tr/td[6]')
    for tag in loc_tags:
        loc.append(tag.text)
    
    
df_funding= pd.DataFrame({"Date":date,"Startup Name":start_name,"Industry/Vertical":industry,"sub-Vertical":sub_vert,"City/Location":loc
                        ,"Investor's Name":invname,"Investment Type":invtype,"Amount(In USD)":amt})
df_funding
```

You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
url="https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2021-12-16&to=2021-12-19&guests=2&page=1"
driver.get(url)
```

```python
links=[]
for link in driver.find_elements_by_xpath("//h2[@class='title title-6']//a"):
    links.append(link.get_attribute("href"))
```

```python
links
```

```python
name=[]
distance=[]
rating=[]
total_review=[]
overall_review=[]
private=[]
dorm=[]
fac=[]
descr=[]

try:
    names=driver.find_elements_by_xpath("//h2[@class='title title-6']/a[1]")
    for n in names[0:2]:
        name.append(n.text)
except:
    name.append("-")

try:
    dist=driver.find_elements_by_xpath("//span[@class='description']")
    for d in dist[0:2]:
        distance.append(d.text)
except:
    distance.append("-")
try:
    rtags=driver.find_elements_by_xpath("//div[@class='rating rating-summary-container big']/div[1]")
    for r in rtags[0:2]:
        rating.append(r.text)
except:
    rating.append("-")
try:
    totalr=driver.find_elements_by_xpath("//div[@class='reviews']")
    for t in totalr[0:2]:
        total_review.append(t.text)
except:
    total_review.append("-")
try:
    overallr=driver.find_elements_by_xpath("//div[@class='keyword']")
    for o in overallr[0:2]:
        overall_review.append(o.text)
except:
    overall_review.append("-")

pprice=driver.find_elements_by_xpath("//div[@class='prices-col']/a[1]/div[1]")
for p in pprice[0:2]:
    try:
        private.append(p.text.split("\n")[1])
    except:
        private.append("-")
try:
    dprice=driver.find_elements_by_xpath("//div[@class='price-col'][2]")
    for d in dprice[0:2]:
        dorm.append(d.text.split("\n")[1])
except:
    dorm.append("-")
    
for i in links[0:2]:
    driver.get(i)        
   
    try:
        description=driver.find_element_by_xpath("//div[@class='content collapse-content']")
        descr.append(description.text)
    except:
        descr.append("-")
        
    try:
        facilities=driver.find_element_by_xpath("//ul[@class='facility-sections']")
        fac.append(facilities.text.replace("\n"," "))
    except:
        fac.append("-")
        
```
