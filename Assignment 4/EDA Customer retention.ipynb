# Exploratory Data Analysis of Customers using Indian E-commerce Platforms

We'll import the datasets that have been stored in 2 separate excel files for EDA process.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("customer_retention_dataset.csv")
df_enc = pd.read_csv("encoded_cust_ret_dataset.csv")
```

Lets clean the column names, removing tabs and whitespaces:

```python
collist=[]
for i in df.columns:
    collist.append(i.strip())
collist
```

Check whether any column has a null value

```python
df.isnull().sum().nlargest(71)
```

As we can see there are no null values are present in the dataset, so we can proceed to check uniques values in each column

```python
df.columns=collist
```

```python
df_enc.columns=collist
```

Lets check the unique values in each dataset, so we know what encoding stands for what

```python
unique_values=[]
for i in df.columns:
    unique_values.append(df[i].unique())
unique_values
    
```

```python
unique_values_enc=[]
for i in df_enc.columns:
    unique_values_enc.append(df_enc[i].unique())
unique_values_enc
    
```

We can drop a lot of columns which wouldn't really help in data analysis or model building:

```python
sns.histplot(df['1Gender of respondent'])
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Gender in the dataset")
plt.show()
```

As we can see, this dataset was collected with a survey where more females participated than males.

```python
plt.figure(figsize=(10,5))
sns.barplot(df['2 How old are you?'].unique(),df['2 How old are you?'].value_counts())
plt.xlabel("Age range")
plt.ylabel("Count")
plt.title("Distribution of Age ranges in the dataset")
plt.show()
```

We can infer from the graph above that the survey was mostly participated by people between the age range of 21-50 years indicating that mostly youth and the working people participated in this survey.. 

```python
plt.figure(figsize=(14,5))
sns.barplot(df['3 Which city do you shop online from?'].unique(),df['3 Which city do you shop online from?'].value_counts(sort=True))
plt.xlabel("Cities in which the survey was conducted")
plt.ylabel("Count")
plt.title("Distribution of Cities in the dataset")
plt.show()
```

```python
df['3 Which city do you shop online from?'].value_counts()
```

We can infer from the graph that all the cities are located in the northern India. Delhi NCR accounts for most of the customers whereas Bulandshahr accounts for the least amount of people who belonged to this city and did the survey

We will see which e commerce platform is the most recommended by the customers, but first we'll have to clean the data in the last column so we can analysze it properly.

```python
l=[]
for i in df['Which of the Indian online retailer would you recommend to a friend?']:
    x = i.split(',')
    for i in x:
        l.append(i.strip())
l
```

```python
list(set(l))           #to check the unique values in the list
```

```python
pd.Series(l).value_counts()
```

```python
plt.figure(figsize=(13,5))
sns.barplot(list(set(l)),pd.Series(l).value_counts(sort=False))
plt.xlabel("Platform Recommended by most users")
plt.ylabel("Number of people")
plt.title("Customer Recommendation ")
plt.show()
```

```python
plt.figure(figsize=(13,5))

sns.barplot(df['43 Shopping on the website gives you the sense of adventure'].unique(),df['43 Shopping on the website gives you the sense of adventure'].value_counts())
plt.xlabel("People feel shopping online is an adventure")
plt.ylabel("Count")
plt.title("Distribution of people's view about shopping online in the dataset")
plt.show()
```

We can conclude from the above graph that more than half of the people feel adventurous when shopping online. They feel a rush internally which can be attributed to online shopping.

```python
plt.figure(figsize=(13,5))

sns.barplot(df['7 How do you access the internet while shopping on-line?'].unique(),df['7 How do you access the internet while shopping on-line?'].value_counts(sort=False))
plt.xlabel("People feel shopping online is an adventure")
plt.ylabel("Count")
plt.title("Distribution of people's view about shopping online in the dataset")
plt.show()
```

We can see that most users access these websites through their mobile internet which indicates that all the ecommerce platforms are being accessed on mobile phones.

```python
plt.figure(figsize=(10,5))
sns.stripplot(x=df['2 How old are you?'],y=df['3 Which city do you shop online from?'])
plt.xlabel("Age group customers belong to")
plt.ylabel("Cities They belong to")
plt.show()
```

We can conclude from the above that most cities lack users from the age range "Less than 20 years" and "51 years and above". We can also see that Delhi NCR has most of the younger ages as the participants.Bulandshahr lacks in almost each age range. From the graph, Solan looks like it has the most users who are below 20 years of age whereas Gurgaon is home to most participants with ages 51 and above . Delhi tops the age group of users in the range 41-50 years.

```python
plt.figure(figsize=(10,5))
sns.stripplot(x=df['1Gender of respondent'],y=df['3 Which city do you shop online from?'])
plt.xlabel("Gender customers belong to")
plt.ylabel("Cities They belong to")
plt.show()
```
